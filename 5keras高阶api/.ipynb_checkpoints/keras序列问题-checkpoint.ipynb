```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers 
```

```python
data = keras.datasets.imdb
```

```python
max_word = 10000
```

```python
(x_train,y_train),(x_test,y_test) = data.load_data(num_words=max_word)
```

```python
x_train.shape,y_train.shape
```

```python
x_train[0]
```

# 文本训练成密集向量

```python
[len(x) for x in x_train]
```

```python
#把文本填充到一样的长度
x_train = keras.preprocessing.sequence.pad_sequences(x_train,300)
x_test = keras.preprocessing.sequence.pad_sequences(x_test,300)
```

```python
y_train
```

```python
demo = 'i am a it'
```

```python
dict((word,demo.split().index(word)) for word in demo.split())
```

```python
model = keras.models.Sequential()
```

```python
model.add(layers.Embedding(10000,50,input_length=300))
```

```python
# model.add(layers.Flatten())
model.add(layers.GlobalAveragePooling1D())
```

```python
model.add(layers.Dense(128,activation='relu'))
```

```python
model.add(layers.Dropout(0.5))
```

```python
model.add(layers.Dense(1,activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(optimizer=keras.optimizers.Adam(lr=0.001),
             loss='binary_crossentropy',
             metrics=['acc'])
```

```python
history = model.fit(x_train,y_train,epochs=15,batch_size=256,validation_data=(x_test,y_test))
```

```python
history.history.keys()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.plot(history.epoch,history.history.get('loss'))
plt.plot(history.epoch,history.history.get('val_loss'))
plt.legend()
```

```python
plt.plot(history.epoch,history.history.get('acc'))
plt.plot(history.epoch,history.history.get('val_acc'))
plt.legend()
```
